```python
import pandas as pd
```

```python
# Read the file
df = pd.read_csv("Procurement KPI Analysis Dataset.csv")
df
```

```python
# Check missing values before dropping
print(df.isnull().sum())
```

```python
# Drop rows that contain at least one missing value
df = df.dropna()
df
```

```python
# Check again after dropping
print(df.isnull().sum())
```

```python
#Show dataset size after cleaning
print("\nDataset shape after cleaning:", df.shape)
```

```python
print(df.describe())
```

```python
print("Missing values after cleaning:")
print(df.isnull().sum())

print("\nZero values after cleaning:")
print((df == 0).sum())
```

```python
# Check for missing values
print("Missing values:\n", df.isnull().sum())

# Check the value to zero
print("\nZero values:\n", (df == 0).sum())

# Duplicate data check
print("\nDuplicate rows:", df.duplicated().sum())
```

```python
#Save a new file after deleting data.
df.to_excel("cleaned_dataset_final.xlsx", index=False)
```
